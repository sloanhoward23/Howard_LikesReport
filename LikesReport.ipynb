This line of code sets "title" as accounts and gives me a list of my last 10 recent likes. 

```python
accounts = [x['title'] for x in mylikes['likes_media_likes']]
accounts[:10]
```

```python
timestamp = [x['string_list_data'][0]['timestamp'] for x in mylikes['likes_media_likes']]
timestamp[:10]
```

Shows the amount of Instagram posts that I have given a like since the creation of my account. 

```python
print(len(accounts), len(timestamp))
```

```python
mylikes_df = pd.DataFrame({'accounts': accounts, 'timestamp': timestamp})
mylikes_df.head()
```

```python
count_mylikes = mylikes_df.groupby('accounts').count()
count_mylikes.head()
```

The list below shows my top 5 most liked accounts since I've had my Instagram. 

```python
count_mylikes.sort_values('timestamp' ,ascending=False).head()
```

Here I am implementing matplotlib and pyplot so I can graph these statistics.

```python
from matplotlib import pyplot as plt
```

Also below are the same statistics from the list above, but shown in a bar graph and a pie graph.

```python
sortcount = count_mylikes.sort_values('timestamp' ,ascending=False)
labels = sortcount['timestamp'].keys()
vals = sortcount['timestamp']
width = .75

fig, ax = plt.subplots(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
ax.bar(labels, vals, width)
ax.set_ylabel('Likes')
plt.xticks(rotation=90, fontsize = 'small')
plt.show()
```

```python
fig2, ax2 = plt.subplots()
ax2.pie(vals, labels = labels)
ax2.axis('equal')
plt.show()
```
